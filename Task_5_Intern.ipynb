```python
import pandas as pd

df = pd.read_csv("TRAIN_1.csv")

df.info()
print()

print(df.describe())
print()

print(df.isnull().sum())
print()

print("Survival counts:")
print(df['Survived'].value_counts())
print()

print("Gender counts:")
print(df['Sex'].value_counts())
print()

print("Passenger class counts:")
print(df['Pclass'].value_counts())

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

sns.histplot(df['Age'].dropna(), bins=30, kde=True)
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

```

```python
sns.countplot(x='Pclass', data=df)
plt.title("Passenger Class Count")
plt.xlabel("Pclass")
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
print()
```

```python
sns.countplot(x='Sex', data=df)
plt.title("Gender Count")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

```python
# Survival by Passenger Class (Third Bivariate Plot)
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Pclass', hue='Survived')
plt.title("Survival by Passenger Class")
plt.xlabel("Pclass")
plt.ylabel("Count")
plt.tight_layout()
plt.subplots_adjust(bottom=0.3)  # Adds more space at the bottom of the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compute the correlation matrix for relevant numerical features
corr = df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].corr()

# Plot the correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix")
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select numerical columns for pairplot
numerical_features = ['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']

# Create a pairplot to visualize relationships
sns.pairplot(df[numerical_features], hue='Survived', palette='coolwarm')
plt.suptitle('Pairplot of Selected Features', y=1.02)
plt.tight_layout()
plt.show()
```
